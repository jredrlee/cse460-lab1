```python
# Perfect sensor
def sense(x):
    return x
```

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

```python
def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    
    #default control policy
    #ux = -sin(t)
    #uy = cos(t)
    #return array([ux, uy])

    #control policy for problem 1
    #ux = 4 *  -sin(t)
    #uy = 2 *  cos(t)
    #return array([ux, uy])
    
    #control policy for problem 2
    rad = 0.523599
    ux = (4 *  -sin(t - rad)) 
    uy = (2 *  cos(t))
    return array([ux, uy])
    
    
```

```python
tf = 7.
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
#x = array([2., 1.])
# initial conditions for problem 1
x = array([7.,2.])


x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)    
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```

```python
grid()
plot(x_log[:,0], x_log[:,1])
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation
from JSAnimation import IPython_display    
from IPython.display import HTML
    


fig, ax = plt.subplots()

def animate(t):
    ax.clear()
    
    # Path
    plot(x_log[:,0], x_log[:,1], 'r--')
    
    # Initial conditions
    plot(x_log[t,0], x_log[t,1], 'bo')
    
    

anim = animation.FuncAnimation(fig, animate, frames=len(time), interval=60)

HTML(anim.to_jshtml())
```
